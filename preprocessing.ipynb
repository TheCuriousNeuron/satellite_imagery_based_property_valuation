```python
# Create a copy for preprocessing
df_clean = df.copy()
```

```python
# Convert date to datetime
df_clean['date'] = pd.to_datetime(df_clean['date'], errors='coerce')
df_clean['year'] = df_clean['date'].dt.year
df_clean['month'] = df_clean['date'].dt.month
df_clean['day_of_week'] = df_clean['date'].dt.dayofweek
print("Date features extracted")
```

```python
# Handle yr_renovated (0 means never renovated)
df_clean['is_renovated'] = (df_clean['yr_renovated'] != 0).astype(int)
df_clean['years_since_renovation'] = np.where(df_clean['yr_renovated'] > 0,df_clean['year'] - df_clean['yr_renovated'],df_clean['year'] - df_clean['yr_built'])
print("Renovation features created")
```

```python
# House age feature
current_year = df_clean['year'].max()
df_clean['house_age'] = current_year - df_clean['yr_built']
print("House age calculated")
```

```python
# Verify sqft_living relationship
df_clean['sqft_check'] = abs(df_clean['sqft_living'] - (df_clean['sqft_above'] + df_clean['sqft_basement']))
inconsistent = (df_clean['sqft_check'] > 1).sum()
print(f"sqft_living consistency check: {inconsistent} inconsistent records")
```

```python
# Create basement binary feature
df_clean['has_basement'] = (df_clean['sqft_basement'] > 0).astype(int)
print("Basement indicator created")
```

```python
# Create size ratios
df_clean['living_lot_ratio'] = df_clean['sqft_living'] / (df_clean['sqft_lot'] + 1)
print("Living to lot ratio calculated")
df_clean['above_living_ratio'] = df_clean['sqft_above'] / (df_clean['sqft_living'] + 1)
print("Above ground ratio calculated")
```

```python
# Neighborhood comparison features
df_clean['living_vs_neighbors'] = df_clean['sqft_living'] / (df_clean['sqft_living15'] + 1)
print("Neighborhood comparison features created")
df_clean['lot_vs_neighbors'] = df_clean['sqft_lot'] / (df_clean['sqft_lot15'] + 1)
```

```python
# Price per sqft
df_clean['price_per_sqft'] = df_clean['price'] / df_clean['sqft_living']
print("Price per sqft calculated")
```

```python
# Total rooms feature
df_clean['total_rooms'] = df_clean['bedrooms'] + df_clean['bathrooms']
print("Total rooms calculated")
```

```python
# Outlier Detection using IQR 
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((data[column] < lower) | (data[column] > upper)).sum()
    return outliers, lower, upper

outlier_cols = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors']
outlier_cols = [col for col in outlier_cols if col in df_clean.columns]

outlier_summary = []
for col in outlier_cols:
    count, lower, upper = detect_outliers_iqr(df_clean, col)
    outlier_summary.append({
        'Column': col,
        'Outlier_Count': count,
        'Outlier_Percentage': round(count/len(df_clean)*100, 2),
    })

outlier_df = pd.DataFrame(outlier_summary)
print(outlier_df)
```

```python
# Geospatial Feature Engineering

# Calculate distance from city center (assuming Seattle: 47.6062° N, 122.3321° W)
city_lat, city_long = 47.6062, -122.3321 
df_clean['dist_from_center'] = np.sqrt((df_clean['lat'] - city_lat)**2 + (df_clean['long'] - city_long)**2) * 111  # Convert to km (approximate)
print("Distance from city center calculated")
    
# Identify potential waterfront areas based on lat/long
# This is a simplified heuristic - actual waterfront detection would need more data
df_clean['near_water_geo'] = ((df_clean['lat'] > 47.5) & (df_clean['lat'] < 47.7) & (df_clean['long'] < -122.2)).astype(int)
print("Geographic waterfront proximity estimated")
```

```python
# Correlation Analysis
numeric_cols = df_clean.select_dtypes(include=[np.number]).columns.tolist()
price_corr = df_clean[numeric_cols].corr()['price'].sort_values(ascending=False)
print("\nTop 15 features correlated with price:")
print(price_corr.head(15))
```

```python
# Visualisation
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
# Price distribution
axes[0, 0].hist(df_clean['price'], bins=50, edgecolor='black', alpha=0.7)
axes[0, 0].set_title('Price Distribution', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Price ($)')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].axvline(df_clean['price'].median(), color='red', linestyle='--', label=f"Median: ${df_clean['price'].median():,.0f}")
axes[0, 0].legend()
# Log-transformed price
axes[0, 1].hist(np.log10(df_clean['price']), bins=50, edgecolor='black', alpha=0.7, color='green')
axes[0, 1].set_title('Log-Transformed Price Distribution', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Log10(Price)')
axes[0, 1].set_ylabel('Frequency')
# Price vs sqft_living
axes[1, 0].scatter(df_clean['sqft_living'], df_clean['price'], alpha=0.3, s=10)
axes[1, 0].set_title('Price vs Living Space', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Square Feet Living')
axes[1, 0].set_ylabel('Price ($)')
# Grade vs Price
grade_price = df_clean.groupby('grade')['price'].median().sort_index()
axes[1, 1].bar(grade_price.index, grade_price.values, edgecolor='black', alpha=0.7, color='coral')
axes[1, 1].set_title('Median Price by Grade', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Grade')
axes[1, 1].set_ylabel('Median Price ($)')
plt.tight_layout()
plt.show()
```

```python
# Geospatial visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))    
# Plot 1: Geographic distribution
scatter = axes[0].scatter(df_clean['long'], df_clean['lat'], c=df_clean['price'], cmap='YlOrRd', alpha=0.5, s=5)
axes[0].set_title('House Prices by Location', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')
plt.colorbar(scatter, ax=axes[0], label='Price ($)')    
# Plot 2: Waterfront comparison
waterfront_price = df_clean.groupby('waterfront')['price'].median()
axes[1].bar(['No Waterfront', 'Waterfront'], waterfront_price.values, color=['skyblue', 'navy'], edgecolor='black', alpha=0.7)
axes[1].set_title('Median Price: Waterfront vs Non-Waterfront', fontsize=14, fontweight='bold')
axes[1].set_ylabel('Median Price ($)')        
# Add value labels
for i, v in enumerate(waterfront_price.values):
    axes[1].text(i, v, f'${v:,.0f}', ha='center', va='bottom', fontweight='bold')    
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
fig, ax = plt.subplots(figsize=(14, 10))
correlation_cols = ['price','bedrooms','bathrooms','sqft_living','sqft_lot','floors','waterfront','view','condition', 'grade', 'sqft_above', 'sqft_basement']
correlation_cols = [col for col in correlation_cols if col in df_clean.columns]
corr_matrix = df_clean[correlation_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0, square=True, linewidths=1, ax=ax)
ax.set_title('Feature Correlation Heatmap', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()
```
